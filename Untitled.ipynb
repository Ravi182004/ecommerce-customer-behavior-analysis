```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("ecommerce_dataset.csv")
```

```python
df['Satisfaction Level'] = df['Satisfaction Level'].fillna(df['Satisfaction Level'].mode()[0])
```

```python
df.duplicated().sum()
```

```python
sns.boxplot(x=df['Total Spend'])
plt.show()
```

```python
sns.boxplot(x=df['Items Purchased'])
plt.show()
```

```python
sns.boxplot(x=df['Average Rating'])
plt.show()
```

```python
le = LabelEncoder()
df['Gender_encoded'] = le.fit_transform(df['Gender'])
df['City_encoded'] = le.fit_transform(df['City'])
df['Membership_encoded'] = le.fit_transform(df['Membership Type'])
df['Satisfaction_encoded'] = le.fit_transform(df['Satisfaction Level'])
```

```python
gender_seg = df.groupby('Gender')[['Total Spend','Items Purchased','Average Rating']].mean().reset_index()
age_seg = df.groupby('Age')[['Total Spend','Items Purchased','Average Rating']].mean().reset_index()
city_seg = df.groupby('City')[['Total Spend','Items Purchased','Average Rating']].mean().reset_index()
membership_seg = df.groupby('Membership Type')[['Total Spend','Items Purchased','Average Rating']].mean().reset_index()
```

```python
df['Churn Risk'] = df['Days Since Last Purchase'].apply(lambda x: 'High' if x>90 else ('Medium' if x>30 else 'Low'))
df['Churn Risk'].value_counts()
```

```python
plt.figure()
sns.countplot(x='Churn Risk', data=df)
plt.show()
```

```python
churn_segment = df.groupby(['Churn Risk','Membership Type'])[['Total Spend','Items Purchased']].mean().reset_index()
churn_segment
```
